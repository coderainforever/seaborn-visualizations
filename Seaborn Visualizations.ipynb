# Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

# Import Data

```python
print(sns.get_dataset_names()) #built-in datasets
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df
```

# Distribution Plots

## Distribution Plot: Single Variable - Histogram Plot

```python
sns.distplot(crash_df['not_distracted'], kde=False, bins=25) # remove kde
```

## Joint Plot: Scatter Plot from two Dist Plots

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')
#sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
#sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde')
```

## KDE Plot

```python
sns.kdeplot(crash_df['alcohol'])
```

## Pair Plot: 
Relationships across entire dataframe's numberical values

```python
# sns.pairplot(crash_df)

tips_df = sns.load_dataset('tips')
#tips_df

sns.pairplot(tips_df, hue='sex', palette="Blues")
```

## Rug Plot

```python
sns.rugplot(tips_df['tip']) 
#single coloumn of data points as sticks. 
#Dense number of lines at places where values are more common
```

## Styling

```python
sns.set_style('ticks')

plt.figure(figsize = (8,4))

sns.set_context('paper', font_scale=1.5)

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')

#sns.despine(left=True, bottom=True)
```

# Categoricla Plots:
Distribition using categorical data

## Bar Plot

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.std) 
# default estimation is based on mean and can be estimated using median, 
# standard deviation, variance and covariance or own function
```

## Count Plot
counting the number of occurances

```python
sns.countplot(x='sex', data=tips_df)
```

## Box Plot

```python
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')

plt.legend(loc=0)
```

## Violin Plot - Comination of Box & KDE

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)

plt.legend(loc=0)
```

## Strip Plot

```python
plt.figure(figsize=(8,5))

sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True )
```

## Swarm Plot

```python
#sns.violinplot(x='day', y='total_bill', data=tips_df)
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='red')
```

# Palettes

```python
plt.figure(figsize=(8,7))
sns.set_style('white')
sns.set_context('talk')
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex', palette='autumn' )
plt.legend(loc=0)
```

# Matrix Plots

## Heatmaps

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)

crash_mx = crash_df.corr()
#crash_mx

sns.heatmap(crash_mx, annot=True, cmap='Blues')
```

```python
flights = sns.load_dataset('flights')

flights = flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(flights, cmap='Blues', linecolor='white', linewidth=1)
```

## Cluster Map

```python
iris = sns.load_dataset('iris')

species = iris.pop('species')

sns.clustermap(iris)
```

```python
sns.clustermap(flights, cmap='Blues', standard_scale=1)
```

## Pair Grids

```python
iris = sns.load_dataset('iris')
#iris_g = sns.PairGrid(iris, hue='species')
#iris_g.map(plt.scatter)
#iris_g.map_diag(plt.hist)
#iris_g.map_offdiag(plt.scatter)
#iris_g.map_upper(plt.scatter)
#iris_g.map_lower(sns.kdeplot)
iris_g = sns.PairGrid(iris, hue='species', 
                      x_vars=['sepal_length', 'sepal_width'],
                      y_vars=['petal_length', 'petal_width'],)
iris_g.map(plt.scatter)
iris_g.add_legend()
```

## Facet Grid

```python
#tips_fg = sns.FacetGrid(tips_df, col='time', row='smoker')
tips_fg = sns.FacetGrid(tips_df, col='time', hue='smoker', height=4, aspect=1.3,
                       col_order=['Dinner', 'Lunch'], palette='Set1')

#tips_fg.map(plt.hist, 'total_bill', bins=8)
#tips_fg.map(plt.scatter, 'total_bill', 'tip')
tips_fg.map(plt.scatter, 'total_bill', 'tip', edgecolor='w')
```

```python
kws = dict(s=50, linewidth=.5, edgecolor='w')
tips_fg = sns.FacetGrid(tips_df, col='sex', hue='smoker', height=4, aspect=1.3,
                       hue_order=['Yes', 'No'],
                       hue_kws=dict(marker=['^', 'v']))
tips_fg.map(plt.scatter,'total_bill', 'tip', **kws)
```

```python
att_df = sns.load_dataset('attention')

att_fg = sns.FacetGrid(att_df, col='subject', col_wrap=5, height=1.5)
att_fg.map(plt.plot, 'solutions', 'score', marker='.')
```
